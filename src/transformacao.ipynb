# Transformação pt 01

-Ajustando os valores das colunas 

'NT_FG', 'NT_CE', 'NT_GER'

```python
colunas = ['CO_IES', 'CO_CATEGAD', 'CO_GRUPO', 'CO_UF_CURSO', 'CO_REGIAO_CURSO' , 'NU_IDADE', 'TP_SEXO', 'NT_FG', 'NT_CE', 'NT_GER' ]

enade = df[colunas]

enade
```

```python
filtro_1 = df['NT_GER'] > 100
filtro_1
enade.loc[filtro_1,'NT_GER'] = df['NT_GER'] / 10
enade
```

```python
filtro_2 = df['NT_CE'] > 100
filtro_2
enade.loc[filtro_2,'NT_CE'] = df['NT_CE'] / 10
enade
```

```python
filtro_3 = df['NT_FG'] > 100
filtro_3
enade.loc[filtro_3,'NT_FG'] = df['NT_FG'] / 10
enade
```

```python
# Criando nova coluna com a Descrição da universidade. 

enade['DESC_PUBLICA'] = ''
enade.loc[
    enade.CO_CATEGAD.isin([1,2,3]),
    'DESC_PUBLICA'
] = 'Publica'
```

```python
enade.loc[
    enade.CO_CATEGAD.isin([4,5]),
    'DESC_PUBLICA'
] = 'Privado'
```

```python
enade.loc[
    enade.CO_CATEGAD.isin([7]),
    'DESC_PUBLICA'
] = 'Especial'
```

```python
# Valores

enade.DESC_PUBLICA.value_counts()
```

```python
# Valores em %
enade.DESC_PUBLICA.value_counts() / enade.shape[0]
enade
```

```python
enade['CO_REGIAO_CURSO'] = enade.CO_REGIAO_CURSO.replace({
    1 : 'Norte',
    2 : 'Nordeste',
    3 : 'Sudeste',
    4 : 'Sul',
    5 : 'Centro Oeste'
})
enade
```

# Salvando dados tratados no modelo csv e no sqlite

```python
# Salvando no SQLite

# Conectando ao banco de dados SQLite
conn = sqlite3.connect('caminho.db')

# Salvando o DataFrame no banco de dados SQLite
enade.to_sql('enade_2019', conn, if_exists='replace', index=False)

# Fechando a conexão com o banco de dados
conn.close()
```

```python
#Salvando em formato CSV

enade.to_csv('caminho_csv', sep=';', decimal= ',', index=False)
```
